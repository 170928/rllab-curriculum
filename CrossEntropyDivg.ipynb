```python
database = ExperimentDatabase("data/crossE/")
```

```python
exps = database._experiments
for exp in exps:
    print exp.params['exp_name']
```

```python
for exp in exps:
    if 'crossE_dvg' in exp.params['exp_name']:
        dvg = exp.params['algo']['dvg']
        lab = "dvg = {}".format(dvg) if dvg else "Baseline"
        plt.plot(exp.progress['AverageDiscountedReturn'], label = lab )
    elif exp.params['exp_name']=='trpo_hopper_1000n':
        lab = "Original TRPO"
        plt.plot(exp.progress['AverageDiscountedReturn'], label = lab )
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
for exp in exps:
    if 'crossE' in exp.params['exp_name']:
        dvg = exp.params['algo']['dvg']
        lab = "dvg = {}".format(dvg) if dvg else "Baseline"
        plt.plot(exp.progress['MeanKLother'], label = lab )
    else:
        lab = "Original TRPO"
        plt.plot(exp.progress['Men'])
plt.yscale('symlog')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
for exp in exps:
    if 'crossE_clipDenom' in exp.params['exp_name'] :
        hyper = exp.params['algo']['clip_denom']
        lab = "clipDenom = {}".format(hyper) if hyper else "Baseline"
        plt.plot(exp.progress['AverageDiscountedReturn'], label = lab )
    elif exp.params['exp_name']=='trpo_hopper_1000n':
        lab = "Original TRPO"
        plt.plot(exp.progress['AverageDiscountedReturn'], label = lab )
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
for exp in exps:
    if 'crossE_clipDenom100' in exp.params['exp_name'] :
        hyper = exp.params['algo']['clip_denom']
        lab = "clipDenom = {}".format(hyper) if hyper else "Baseline"
        plt.plot(exp.progress['AverageDiscountedReturn'], label = lab )
        plt.plot(exp.progress['MeanKLother'], label = "MeanKL_other " + lab)
    elif exp.params['exp_name']=='trpo_hopper_1000n':
        lab = "Original TRPO"
        plt.plot(exp.progress['AverageDiscountedReturn'], label = lab )
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
